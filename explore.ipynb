```python
import json
import numpy as np
import pandas as pd

# Columns whose float/int values are Unix-ms timestamps
TIMESTAMP_COLS = {c for c in gdf.columns if
    any(kw in c for kw in ('DateTime', 'CreateDate', 'DateCurrent',
                            'PolygonDateTime', 'ApprovedDate',
                            'PublishDate', '_dt'))}

stats = {}

for col in gdf.columns:
    if col == 'geometry':
        stats[col] = {'dtype': 'geometry', 'kind': 'geometry'}
        continue

    s = gdf[col]
    dtype = str(s.dtype)
    n = len(s)
    null_count = int(s.isna().sum())
    non_null = n - null_count

    entry = {
        'dtype': dtype,
        'count': n,
        'non_null': non_null,
        'null_count': null_count,
        'null_pct': round(null_count / n * 100, 1),
    }

    if col in TIMESTAMP_COLS and pd.api.types.is_numeric_dtype(s):
        try:
            unit = 'ms' if s.dropna().abs().median() > 1e10 else 's'
            dt = pd.to_datetime(s, unit=unit, errors='coerce')
            valid = dt.dropna()
            if len(valid):
                entry['kind'] = 'datetime'
                entry['min'] = str(valid.min())[:19]
                entry['max'] = str(valid.max())[:19]
            else:
                entry['kind'] = 'numeric'
        except Exception:
            entry['kind'] = 'numeric'

    elif pd.api.types.is_numeric_dtype(s):
        valid = s.dropna()
        entry['kind'] = 'numeric'
        if len(valid):
            entry['mean']   = round(float(valid.mean()), 3)
            entry['median'] = round(float(valid.median()), 3)
            entry['min']    = round(float(valid.min()), 3)
            entry['max']    = round(float(valid.max()), 3)
            entry['std']    = round(float(valid.std()), 3)

    else:
        entry['kind'] = 'categorical'
        entry['unique'] = int(s.nunique(dropna=True))
        vc = s.value_counts(dropna=True).head(7)
        entry['top_values'] = [[str(k), int(v)] for k, v in vc.items()]
        ex = s.dropna()
        entry['example'] = str(ex.iloc[0]) if len(ex) else None

    stats[col] = entry

with open('column_stats.js', 'w', encoding='utf-8') as f:
    f.write('const COLUMN_STATS = ')
    json.dump(stats, f, indent=2, default=str)
    f.write(';')

print(f'column_stats.js written — {len(stats)} columns')
```
